```python
from Scheduler import Scheduler

sch = Scheduler()
testSch = GossipGraph([("a", "b"), ("c", "b")])

testSch.add_rule("a", lambda g: not g.is_familiar_with("a", "C"), ('a', 'b'))
testSch.add_rule("b", lambda g: g.is_familiar_with("b", "A") and not g.is_familiar_with("b", "C"), ('b', 'c'))

sch.run(testSch)
testSch.draw()
```
